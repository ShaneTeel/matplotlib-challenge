```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D  # noqa: F401 (needed for 3D plotting)
```

```python
titanic_df = pd.read_csv('plt_code_along/data/titanic_survival.csv')
titanic_df.head()
```

```python
# Drop rows with missing values in the columns we want to plot
df = titanic_df[['age', 'fare', 'pclass']].dropna()
    
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')
    
# Scatter plot
ax.scatter(df['age'], df['fare'], df['pclass'], c=df['pclass'], cmap='viridis', s=60, alpha=0.7)
    
ax.set_xlabel('Age')
ax.set_ylabel('Fare')
ax.set_zlabel('Pclass')
ax.set_title('Titanic: Age vs Fare vs Pclass (3D Scatter)')
    
plt.show()
```
